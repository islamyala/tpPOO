```python
# Import the Pandas library with the alias "pd"
import pandas as pd

# Read the CSV file "titanic-passengers.csv" with a semicolon (;) as the field separator
df = pd.read_csv("titanic-passengers.csv", sep=';')

# Display the first five rows of the DataFrame for initial data inspection
df.head()

```

```python
# Get and display essential information about the DataFrame
df.info()

```

```python
# Count the number of missing values (NaN) in each column of the DataFrame and store the results in a variable
missing_values = df.isnull().sum()
```

```python
# Print the counts of unique values in the "Age" column of the DataFrame
print(df["Age"].value_counts())
```

```python
# Fill missing values in the "Age" column with the value 24 and apply the change to the DataFrame in-place
df["Age"].fillna(24, inplace=True)
```

```python
# Print the counts of unique values in the "Cabin" column of the DataFrame
print(df["Cabin"].value_counts())
```

```python
# Fill missing values in the "Cabin" column with the value "G6" and apply the change to the DataFrame in-place
df["Cabin"].fillna("G6", inplace=True)
```

```python
# Print the counts of unique values in the "Embarked" column of the DataFrame
print(df["Embarked"].value_counts())
```

```python
# Fill missing values in the "Embarked" column with the value "S" and apply the change to the DataFrame in-place
df["Embarked"].fillna("S", inplace=True)
```

```python
# Import necessary libraries: Seaborn for data visualization, Matplotlib for plotting, and Pandas for data manipulation
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define a function to plot a correlation map for a DataFrame
def plot_correlation_map(df):
    # Calculate the correlation matrix
    corr = df.corr()
    
    # Create a subplots figure with a specific size
    s, ax = plt.subplots(figsize=(12, 10))
    
    # Define a color map for the heatmap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    
    # Create a heatmap of the correlation matrix
    s = sns.heatmap(
        corr,
        cmap=cmap,
        square=True,
        cbar_kws={'shrink': 0.9},
        ax=ax,
        annot=True,
        annot_kws={'fontsize': 12}
    )

# Call the function to plot the correlation map for the DataFrame
plot_correlation_map(df)

# Display the plot
plt.show()

```
